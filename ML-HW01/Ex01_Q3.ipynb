```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
```

#### Reading the dataset

```python
dataset = pd.read_csv('./ML-HW01/dataset3/test.csv')
dataset.head()
```

```python
dataset.info()
```

```python
encoder = preprocessing.LabelEncoder()
dataset['gender'] = encoder.fit_transform(dataset.loc[:, 'gender'])
dataset['smoker'] = encoder.fit_transform(dataset.loc[:, 'smoker'])

dataset.head()
```

```python
dataset.loc[:, 'region'].unique()
```

```python
pd.get_dummies(dataset.loc[:, 'region'])
region = pd.get_dummies(dataset['region'], prefix='region_')
dataset = pd.merge(
    left=dataset,
    right=region,
    left_index=True,
    right_index=True,)

dataset.pop('region')
dataset['charges'] = dataset.pop('charges')

dataset.head()
```
